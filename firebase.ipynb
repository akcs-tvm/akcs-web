```python
import csv
import firebase_admin
from firebase_admin import credentials, firestore

# Initialize Firestore
cred = credentials.Certificate("./akcs.json")  # Replace with your service account key
app =firebase_admin.initialize_app(cred)
db = firestore.client()
```

```python
import pandas as pd
csv_file = "./src/data/data-2026.xlsx"  # Update with your file name
df = pd.read_excel(csv_file)
```

```python
df["Mobile No."] = df["Mobile No."].fillna(0)
```

```python
collection_name = "csp-tvm-2026"

# Upload each row as a Firestore document
for index, row in df.iterrows():
    doc_id = str(row["CSP Memb No."])  # Use 'number' as the document ID
    data = {
        "name": row["Name"],
        "nominee": row["Nominee"],
        "unit": row["Unit"],
        "place": row["Place"],
        "phone": int(row["Mobile No."]),  
    }
    db.collection(collection_name).document(doc_id).set(data)
```

```python
COLLECTION_NAME = "csp-tvm-2026"

# üîπ Function to Query by Document ID (CSP Number)
def query_by_document_id(csp_number):
    doc_ref = db.collection(COLLECTION_NAME).document(str(csp_number))
    doc = doc_ref.get()

    if doc.exists:
        print(f"‚úÖ CSP Number {csp_number} found in Firestore!")
        print("üìå User Data:", doc.to_dict())
    else:
        print(f"‚ùå No document found for CSP Number {csp_number}")

# üîπ Function to Query by Place
def query_by_place(place):
    docs = db.collection(COLLECTION_NAME).where("place", "==", place).stream()
    print(f"üìç Users from '{place}':")
    for doc in docs:
        print(f" - {doc.id} ‚Üí {doc.to_dict()}")

# üîπ Function to Query by Name
def query_by_name(name):
    docs = db.collection(COLLECTION_NAME).where("name", "==", name).stream()
    print(f"üîç Users with Name '{name}':")
    for doc in docs:
        print(f" - {doc.id} ‚Üí {doc.to_dict()}")

# üîπ Function to Query by Multiple Conditions (Example: Place & Unit)
def query_by_place_and_unit(place, unit):
    docs = db.collection(COLLECTION_NAME).where("place", "==", place).where("unit", "==", unit).stream()
    print(f"üìç Users from '{place}' in Unit '{unit}':")
    for doc in docs:
        print(f" - {doc.id} ‚Üí {doc.to_dict()}")

# üîπ Function to List All Documents
def list_all_documents():
    docs = db.collection(COLLECTION_NAME).stream()
    print("üìã Listing all Firestore Documents:")
    for doc in docs:
        print(f" - {doc.id} ‚Üí {doc.to_dict()}")

# ‚úÖ Run Queries
query_by_document_id(101)               # Check if CSP Number (Document ID) 101 exists
query_by_place("MANACAUD")              # Query by Place
query_by_name("ROHINI")                 # Query by Name
query_by_place_and_unit("MANACAUD", "TVDC")  # Query by Multiple Conditions
list_all_documents()                     
```
